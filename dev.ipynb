# Data Descriptions

## 1. Global Land Temperatures By Country

### What are the observations (rows) and the attributes (columns)?

**Observations:** Each row represents a temperature measurement for a specific country on a given date.

**Attributes:**
* `dt`: Date of the temperature measurement
* `AverageTemperature`: Average temperature for the country in Celsius
* `AverageTemperatureUncertainty`: The 95% confidence interval around the average temperature measurement
* `Country`: Name of the country where the temperature was recorded

### Why was this dataset created?

This dataset was created to provide a comprehensive and reliable source of country-level temperature data for studying climate trends across different geographical regions. It aims to enable analysis of temperature patterns and changes at a national level, contributing to the understanding of climate change's varying impacts across different parts of the world.

### Who funded the creation of the dataset?

The dataset was created by Berkeley Earth, which is affiliated with Lawrence Berkeley National Laboratory. The Berkeley Earth Surface Temperature Study combined data from 16 pre-existing archives, containing 1.6 billion temperature reports.

### What processes might have influenced what data was observed and recorded and what was not?

#### Historical measurement methods:
* Early data collection using mercury thermometers
* Variations in measurement timing affecting readings

#### Data collection consistency:
* Different measurement techniques across time periods
* Varying geographical coverage of weather stations
* Changes in observation frequency and timing

### What preprocessing was done, and how did the data come to be in the form that you are using?

Berkeley Earth conducted several preprocessing steps:
1. Combination of data from 16 different archives
2. Application of methods to include shorter time series observations
3. Standardization of measurements across different time periods
4. Statistical processing to calculate global averages and confidence intervals
5. Creation of consistent temperature series despite changes in measurement methods

### If people were involved, were they aware of the data collection and if so, what purpose did they expect the data to be used for?

Berkeley Earth staff specifically collected and processed this data for climate change research and public understanding.

### Where can your raw source data be found?
The raw data is hosted on Github in our data folder and can be accessed through Kaggle at: [https://www.kaggle.com/datasets/berkeleyearth/climate-change-earth-surface-temperature-data](https://www.kaggle.com/datasets/berkeleyearth/climate-change-earth-surface-temperature-data).

# 2. Political Regime Data

### What are the observations (rows) and the attributes (columns)?

**Observations:** Each row represents a country's political regime classification for a specific year.

**Attributes:**
* `Entity`: Name of the country
* `Code`: Three-letter country code
* `Year`: Year of observation
* `Political regime`: Numerical score representing the type of political regime (0-3)
- 0: Closed autocracies
- 1: Electoral autocracies  
- 2: Electoral democracies
- 3: Liberal democracies

### Why was this dataset created?

This dataset was created to classify and track political regimes across countries and time. It aims to provide a standardized way to analyze democratic development and political transitions globally, using clear criteria to distinguish between different types of political systems.

### Who funded the creation of the dataset?

The dataset was created by Our World in Data, based on the Regimes of the World classification developed by political scientists Anna Lührmann, Marcus Tannenberg and Staffan Lindberg. Our World in Data is a non-profit organization that publishes research and data on global issues.

### What processes might have influenced what data was observed and recorded and what was not?

#### Political assessment challenges:
* Difficulty in objectively measuring democratic freedoms
* Potential political bias in regime classification
* Changes in definition of democracy over time

#### Data collection constraints:
* Limited information from closed political systems
* Varying quality of historical political records
* Complex political transitions that may not fit cleanly into categories

### What preprocessing was done, and how did the data come to be in the form that you are using?

The dataset preparation involved:
1. Assessment of countries against standardized criteria for regime classification
2. Conversion of qualitative political assessments into numerical scores
3. Standardization of country names and codes
4. Organization of data into consistent yearly observations
5. Integration of historical records and contemporary assessments

### If people were involved, were they aware of the data collection and if so, what purpose did they expect the data to be used for?

The data was collected and analyzed by political scientists and researchers who were aware their work would be used for:
* Academic research on democracy and political systems
* Policy analysis and development
* Public education about global political systems
* Cross-country comparisons of political development

### Where can your raw source data be found?

The raw data is hosted on Github in our data folder and can be accessed through Our World in Data at: [https://ourworldindata.org/grapher/political-regime](https://ourworldindata.org/grapher/political-regime).

Next, we clean each dataframe one by one – dropping unnecessary columns, dealing with missing values and reorganising them to better prepare us for our future analysis. 

#### Cleaning `temperature_df`

```python
# Preview the df
# temperature_df.head(100)

# Identify numeric columns  
numeric_columns = temperature_df.select_dtypes(include = [np.number]).columns

# Replace missing values with nans and then linearly interpolate to fill them (TODO: find better interpolation mechanisms)
temperature_df.replace('', np.nan, inplace=True)

# Only interpolate numeric columns 
temperature_df[numeric_columns] = temperature_df[numeric_columns].interpolate()

# Convert the dates in the dataframe to datetime type
temperature_df['dt'] = pd.to_datetime(temperature_df['dt'])
temperature_df.rename(columns={'dt':'Date'}, inplace=True)

# Filter for relevant countries
temperature_df = temperature_df[temperature_df['Country'].isin(UN_recognised_countries)].reset_index(drop=True)
temperature_df.head(100)
```

```python
# Create new data set with temperature averages grouped by country by year 
# Right now, we are just taking the average of the temperatures, but this is not 
# completely sound because we have missing data for many months which skews the averages for those years. 
# TODO: use some sort of interpolation for those months (and drop for years with too few datapoints)

yearly_temperature_df = temperature_df.copy()
yearly_temperature_df['Year']=temperature_df['Date'].dt.year

yearly_temperature_df = yearly_temperature_df.groupby(['Country', 'Year']).agg({
    'AverageTemperature': 'mean',
    'AverageTemperatureUncertainty': 'mean'}).reset_index()

yearly_temperature_df.head(100)
```

#### Cleaning `political_regime_df`


```python
# Drop rows with missing values
political_regime_df.replace('', np.nan, inplace=True)
political_regime_df.dropna(inplace=True)

# drop code column
political_regime_df = political_regime_df.drop(columns=['Code'])

# filter data to be after a certain year according to the relevant time period that we determined (1960)
political_regime_df = political_regime_df[political_regime_df['Year'] >= 1960]

# drop any duplicates
political_regime_df = political_regime_df.drop_duplicates()

# check for any missing values
missing_values = political_regime_df.isnull().sum()

# drop any non-country entries
political_regime_df = political_regime_df[political_regime_df['Entity'].isin(UN_recognised_countries)]
political_regime_df.reset_index(drop=True, inplace=True)

# rename columns
political_regime_df.rename(columns={'Entity':'Country', 'Political regime':'PoliticalRegime'}, inplace=True)

political_regime_df.head(100)
```

#### Cleaning `population_df`

```python
# Identify numeric columns  
numeric_columns = population_df.select_dtypes(include = [np.number]).columns

# Interpolate rows with missing values 
population_df.replace('', np.nan, inplace=True)

# Only interpolate numeric columns 
population_df[numeric_columns] = population_df[numeric_columns].interpolate()

# Drop unnecessary columns and  (TODO: years we aren't considering)
population_df = population_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'])
population_df = population_df.rename(columns={'Country Name': 'Country'})

# Convert the df from wide format to long format
population_df = pd.melt(
    population_df,
    id_vars = ['Country'],
    var_name = 'Year',
    value_name = 'Population'
)

# Filter out non-countries
population_df = population_df[population_df['Country'].isin(UN_recognised_countries)]

# Order by country
population_df.sort_values(by='Country', inplace=True)
population_df.reset_index(drop=True, inplace=True)

population_df.head(100)
```

#### Cleaning `GDP_df`

```python
# Identify numeric columns  
numeric_columns = GDP_df.select_dtypes(include = [np.number]).columns

# Interpolate rows with missing values  
GDP_df.replace('', np.nan, inplace=True)

# Only interpolate numeric columns
GDP_df[numeric_columns] = GDP_df[numeric_columns].interpolate()

# Drop unnecessary columns (`Country Code`, `Indicator Name`, `Indicator Code` and years we aren't considering)
columns_to_drop = ['Country Code', 'Indicator Name', 'Indicator Code']
GDP_df = GDP_df.drop(columns = columns_to_drop)

# Convert the df from wide format to long format
GDP_df = pd.melt(
    GDP_df,
    id_vars = ['Country Name'],
    var_name = 'Year',
    value_name = 'GDP'
)   

GDP_df.rename(columns={'Country Name':'Country'}, inplace=True)

# Filter out non-countries
GDP_df = GDP_df[GDP_df['Country'].isin(UN_recognised_countries)]

# Order by country
GDP_df.sort_values(by='Country', inplace=True)
GDP_df.reset_index(drop=True, inplace=True)


GDP_df.head(100)
```

#### Cleaning `climate_policy_df`

```python
# Determine how many unique policy_objectives there are 
climate_policy_df['policy_objective'].nunique()
# There are 84 unique policy_objectives 

# This shows all the combinations of unique policy_objectives 
# Could create a multi-label binarization for this? 

# Deal with the dates — standardize the format
climate_policy_df['decision_date'] = pd.to_datetime(climate_policy_df['decision_date'], format='%Y')

# Ensure we only include rows after 1960 
climate_policy_df = climate_policy_df[climate_policy_df['decision_date']>='1960-01-01']

# Remove unnecessary columns that 
# complicate data, may be empty, or may be too complicated for the work or
# not provide useful information
columns_to_remove = ['country_iso', 
                     'country_update', 'policy_name',
                     'policy_title', 'jurisdiction', 
                     'supranational_region', 'subnational_region', 
                     'policy_city_or_local', 'policy_description', 
                     'stringency', 'policy_status', 'start_date', 
                     'end_date', 'high_impact', 'reference', 
                     'last_update', 'impact_indicators.comments', 
                     'impact_indicators.name', 'impact_indicators.value', 
                     'impact_indicators.base_year', 'impact_indicators.target_year']
climate_policy_df = climate_policy_df.drop(columns = columns_to_remove)

# Rename columns
climate_policy_df.rename(columns={'country':'Country'}, inplace=True)

# Rename United States entries
climate_policy_df.loc[climate_policy_df['Country'] == 'United States of America', 'Country'] = 'United States'

# drop any non-country entries
climate_policy_df = climate_policy_df[climate_policy_df['Country'].isin(UN_recognised_countries)]

# Order by country
climate_policy_df.sort_values(by='Country', inplace=True)
climate_policy_df.reset_index(drop=True, inplace=True)

climate_policy_df.head()

# Is there a way to create dummy variables for the policy objectives?
```

After looking at general trends, we now analyze a linear regression model to see how well past temperature data between 1960 and 1995 can predict test data, post 1995. We then construct a residual plot to check the model's performance. 

```python
early_data = yearly_temperature_df[(yearly_temperature_df['Year']<=2000) & (yearly_temperature_df['Year']>=1960)]
later_data = yearly_temperature_df[yearly_temperature_df['Year']>2000]

#training model 
X_train = np.array(early_data[['Year']])
Y_train = np.array(early_data[['AverageTemperature']])

#testing model
X_test = np.array(later_data[['Year']])
Y_test = np.array(later_data[['AverageTemperature']])

#fit the model on the early data 
temperature_model = LinearRegression().fit(X_train, Y_train)

#predictions on the later data post 1995
Y_pred_train = temperature_model.predict(X_train)
Y_pred_test = temperature_model.predict(X_test)

mae_train = mean_absolute_error(Y_train, Y_pred_train)
mae_test = mean_absolute_error(Y_test, Y_pred_test)

print(f"Train MAE: {round(mae_train, 2)}")
print(f"Test MAE: {round(mae_test, 2)}")

#calculating and plotting residuals to see fit 
residuals = Y_test - Y_pred_test
plt.figure(figsize=(10, 6))

plt.scatter(Y_pred_test, residuals)
plt.axhline(0, color = 'blue', linestyle = 'dashed')
plt.xlabel('Predicted Temperature')
plt.ylabel('Residuals')
plt.title('Residual Plot')

plt.show()
```

Even though the Mean Absolute Errors are similar for the train and test data, the residual plot shows that the data is not uniformly distributed around 0, indicating that the model may be inadequaltely fitting the relationship between year and temperature. For further analysis, this indicates that we should transform our data to eliminate these issues.
